```python
import pandas
```

```python
db = pandas.read_csv('data.csv')
```

```python
y = db['marks']
```

```python
x = db['hrs']
```

```python
x = x.values.reshape(-1,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mind = LinearRegression()
```

```python
mind.fit(x, y)
```

```python
a = mind.predict([[6]])
 
```

```python
mind.coef_
```

```python
import joblib
```

```python
joblib.dump(mind,'marks_model')
```

```python
import joblib
from flask import Flask

app  = Flask(__name__)


model = joblib.load("marks_model")
#hrs = int(input("enter the number of marks you studied"))
@app.route("/marks/<hrs>")
def marks(hrs):
    p = model.predict([[hrs]])

    print(p)
    s = p[0][0]
    return str(s)

app.run(port=80, host='0.0.0.0')
```
